```python
# Cell 15 - Save Cleaned Dataset
df.to_csv('cleaned_skill_data.csv', index=False)
print("Cleaned dataset saved! ✅")
print("Shape:", df.shape)
print("Ready for Power BI! 🎉")
```

```python
# Cell 16 - Build Skill Survival Probability Score
df['survival_score'] = (
    (df['salary_usd'] / df['salary_usd'].max()) * 35 +
    (1 - df['automation_risk_score']) * 30 +
    (1 - df['ai_intensity_score']) * 20 +
    ((df['salary_change_vs_prev_year_percent'] -
      df['salary_change_vs_prev_year_percent'].min()) /
     (df['salary_change_vs_prev_year_percent'].max() -
      df['salary_change_vs_prev_year_percent'].min())) * 15
)

print("Survival Score calculated! ✅")
print(df['survival_score'].describe())
```

```python
# Cell 17 - Classify Skills into 4 Tiers
def classify_skill(score):
    if score >= 75:
        return 'Future-Proof'
    elif score >= 50:
        return 'Stable'
    elif score >= 25:
        return 'At-Risk'
    else:
        return 'Likely Obsolete'

df['skill_category'] = df['survival_score'].apply(classify_skill)

print("Skill Classification done! ✅")
print("\nClassification Summary:")
print(df['skill_category'].value_counts())
```

```python
# Cell 18 - Visualize Skill Classification
plt.figure(figsize=(8,5))
colors = ['#FF6B6B', '#FFD93D', '#6BCB77', '#4D96FF']
df['skill_category'].value_counts().plot(
    kind='bar',
    color=colors,
    edgecolor='white'
)
plt.title('Skill Survival Classification')
plt.xlabel('Category')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('skill_classification.png')
plt.show()
print("Chart 8 done! ✅")
```

```python
# Cell 19 - Top 10 Future Proof Jobs
plt.figure(figsize=(12,6))
future_proof = df[df['skill_category'] == 'Future-Proof'][
    ['job_title', 'industry', 'survival_score',
     'automation_risk_score', 'salary_usd']
].sort_values('survival_score', ascending=False).head(10)

sns.barplot(x='survival_score', y='job_title',
            data=future_proof, palette='Greens_r')
plt.title('Top 10 Future-Proof Jobs by Survival Score')
plt.xlabel('Survival Score')
plt.ylabel('Job Title')
plt.tight_layout()
plt.savefig('top_future_proof_jobs.png')
plt.show()
print(future_proof[['job_title','industry','survival_score','salary_usd']])
print("Chart 9 done! ✅")
```
